```python
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord

# Define the tags
twin_strep = "WSHPQFEKGGGSGGGSGGSAWSHPQFEK"
tev_site = "ENLYFQG"

# Combine tags
full_tag = twin_strep + tev_site

# Your signal peptide (tPA)
signal_peptide = "MDAMKRGLCCVLLLCGAVFVS"

# Input and output files
input_file = "best5.fasta"
output_file = "best5_tagged.fasta"

# Process each sequence
tagged_records = []

for record in SeqIO.parse(input_file, "fasta"):
    seq = str(record.seq)
    
    if not seq.startswith(signal_peptide):
        raise ValueError(f"Sequence '{record.id}' does not start with expected signal peptide!")

    # Slice after signal peptide
    signal_length = len(signal_peptide)
    rest_of_vaccine = seq[signal_length:]

    # Rebuild full sequence: signal peptide + tag + TEV + vaccine
    final_seq = signal_peptide + full_tag + rest_of_vaccine

    new_record = SeqRecord(
        Seq(final_seq),
        id=record.id + "_Ntagged",
        description="SignalPeptide-TwinStrep-TEV-Vaccine"
    )
    tagged_records.append(new_record)

# Save updated sequences
SeqIO.write(tagged_records, output_file, "fasta")
print(f"✅ N-terminal tags inserted. Output saved to '{output_file}'")
```
